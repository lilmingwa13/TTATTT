```python
def last_occurrence(p, kitu):
    for i in range(len(p) - 1, -1, -1):
        if kitu == p[i]:
            return i
    return -1

def xu_ly(t, p):
    m, n = len(p), len(t)
    i, j = m - 1, m - 1
    check = 0
    sopheptinh = 0

    while i < n:
        checkp = 0
        while checkp != m:
            if t[i] == p[j]:
                i -= 1
                j -= 1
                checkp += 1
                sopheptinh += 1
            else:
                last_occ = last_occurrence(p, t[i])
                i += m - min(j, 1 + last_occ)
                j = m - 1
                sopheptinh += 1
                break

        if checkp == m:
            check += 1
            print(f"\nKQ: Vi tri {i + 1}")
            break
            

    if check == 0:
        print("\nKQ: Khong co chuoi can tim")
    else:
        print(f"\nSo phep tinh la: {sopheptinh}")
if __name__ == "__main__":
    t = input("Nhap doan van ban T: ")
    p = input("Nhap chuoi P can tim: ")
    xu_ly(t, p)
```

```python
def failureFunction(p):
    f = [-1, 0]
    for j in range(2, len(p)):
        i = f[j - 1]
        while i >= 0 and p[i] != p[j - 1]:
            i = f[i]
        f.append(i + 1)
    return f

def xuLy(t, p, f):
    i = 0
    j = 0
    check = 0
    sopheptinhlap = 0
    while i < len(t) - len(p):
        checkp = j
        k = i + j
        while checkp != len(p):
            if t[k] == p[j]:
                k += 1
                j += 1
                checkp += 1
                sopheptinhlap += 1
            else:
                i = i + j - f[j]
                j = 0 if f[j] == -1 else f[j]
                sopheptinhlap += 1
                break
        if checkp == len(p):
            print(f"\nKQ: Vi tri {i}")
            check = 1
            j = 0
            i = i + len(p)
    if check == 0:
        print("\nKQ: Khong co chuoi can tim")
    return sopheptinhlap

if __name__ == "__main__":
    t = input("Nhap doan van ban T: ")
    p = input("Nhap chuoi P can tim: ")
    print(f"VÄn báº£n: {t}")
    print(f"Chuá»i cáº§n kiá»m tra: {p}")
    f = failureFunction(p)
    sopheptinhlap = xuLy(t, p, f)
    print(f"\nSo phep tinh lap la: {sopheptinhlap}")
```
